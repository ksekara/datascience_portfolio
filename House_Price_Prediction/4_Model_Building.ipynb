# Machine Learning Model Build
In this notebook we will create the machine learning model using with pre selected features.

```python
# to handle datasets
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
%matplotlib inline

# to build the models
from sklearn.linear_model import Lasso

# to evaluate the models
from sklearn.metrics import mean_squared_error
from math import sqrt

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
# Load data set

X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')

X_train.head()
```

```python
# separate target variable
y_train = X_train['SalePrice']
y_test = X_test['SalePrice']
```

```python
# laod the selected features we have saved during feature selection stage
features = pd.read_csv('selected_features.csv')
features = [x for x in features['0']]
features
```

```python
# reduce the variables in train and test dataset
X_train = X_train[features]
X_test = X_test[features]
X_train.shape, X_test.shape
```

## Model Training

```python
lin_model = Lasso(alpha=0.005, random_state=0)
lin_model.fit(X_train, y_train)
```

## Model Evaluation

```python
# we will evaluate our model performance using mean squared error  and root of mean squared error
pred = lin_model.predict(X_train)
print('linear train mse: {:.2f}'.format(mean_squared_error(np.exp(y_train), np.exp(pred))))
print('linear train rmse: {:.2f}'.format(sqrt(mean_squared_error(np.exp(y_train), np.exp(pred)))))
print()
pred = lin_model.predict(X_test)
print('linear test mse: {:.2f}'.format(mean_squared_error(np.exp(y_test), np.exp(pred))))
print('linear test rmse: {:.2f}'.format(sqrt(mean_squared_error(np.exp(y_test), np.exp(pred)))))
print()
print('Average house price:{:.2f} '.format(np.exp(y_train).median()))
```

```python
# let's evaluate our predictions respect to the original price
plt.scatter(y_test, lin_model.predict(X_test))
plt.xlabel('True House Price')
plt.ylabel('Predicted House Price')
plt.title('Evaluation of Lasso Predictions')
```

We can see that our model is doing good job predicting house price.

```python
# let's explore distribution of errors
errors = y_test - lin_model.predict(X_test)
errors.hist(bins=15)
```

The distirbution of errors follow gaussian distribution. This also a good indication that our model perform better.

## Feature Importance

```python
# Finally, just for fun, let's look at the feature importance

importance = pd.Series(np.abs(lin_model.coef_.ravel()))
importance.index = features
importance.sort_values(inplace=True, ascending=False)
importance.plot.bar(figsize=(18,6))
plt.ylabel('Lasso Coefficients')
plt.title('Feature Importance')
```
