# Feature Engineering

This contains the steps taken to feature engineer the houe price dataset.


Objective of feature engineering is to tackle the following:

1. Missing Values
2. Temporal variables
3. Non-Gaussian distributed variables
4. Categorical variables: remove rare labels
5. Categorical variables: convert strings to numbers
5. Standarise the values of the variables to the same range


### 1. Dataset Loading and Import of Required Libries

```python
# to handle datasets
import pandas as pd
import numpy as np

# for plotting graphs
import matplotlib.pyplot as plt
%matplotlib inline

# to divide train and test set
from sklearn.model_selection import train_test_split

# feature scaling
from sklearn.preprocessing import MinMaxScaler

# to display all the columns of the dataframe in the notebook without truncatin
pd.pandas.set_option('display.max_columns', None)
```

```python
# Load dataset
data = pd.read_csv('house_pricing.csv')
print(data.shape)
data.head()
```

### 2. Split dataset into train and test data

To avoid the over-fitting model, it is import to separate out data into training and test sets. This is done by randomly, so inorder to keep reproducibility we need to set seed. 

```python
# split the data into train and test set

X_train, X_test, y_train, y_test = train_test_split(data, data.SalePrice, test_size=0.1, random_state=0)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 3. Missing Values



#### 3.1 Categorical Variables
For categorical variables, we will fill missing information by adding an additional category "missing".

```python
# make list of categorical variable that contains missing values
vars_with_na = [var for var in data.columns if X_train[var].isnull().sum() > 1 and X_train[var].dtypes == 'O']

#print the variable name and percentage of missing values
for var in vars_with_na:
    print(var, np.round(X_train[var].isnull().mean() * 100, 3), '% missing values')
```

```python
# function to replae NA in categorical variable 
def fill_categorical_na(df, cat_var):
    """
    df-> dataframe
    cat_var-> categorical variables with missing values(list of strings)
    """
    X = df.copy()
    X[cat_var] = df[cat_var].fillna('Missing')
    return X
```

```python
# replace missing values with the "Missing"
X_train = fill_categorical_na(X_train, vars_with_na)
X_test = fill_categorical_na(X_test, vars_with_na)
```

```python
# to confirm there is no missing values in categorical variables
print(X_train[vars_with_na].isnull().sum())
print()
print(X_test[vars_with_na].isnull().sum())
```

#### 3.2 Numerical Variables
For numerical variables we have to add additions statics captures from the variable data. We will use mode to replace na values.

```python
# make list of numerical variable that contains missing values
vars_with_na = [var for var in data.columns if X_train[var].isnull().sum() > 1 and X_train[var].dtypes != 'O']

#print the variable name and percentage of missing values
for var in vars_with_na:
    print(var, np.round(X_train[var].isnull().mean() * 100, 3), '% missing values')
```

```python
# replace missing values
for var in vars_with_na:
    
    # calcualte the mode
    mode_val = X_train[var].mode()[0]
    
    #train
    # adding binary flags for missing data
    X_train[var+'_na'] = np.where(X_train[var].isnull(), 1, 0)
    X_train[var].fillna(mode_val, inplace = True)
    
    #test
    # adding binary flags for missing data
    X_test[var+'_na'] = np.where(X_test[var].isnull(), 1, 0)
    X_test[var].fillna(mode_val, inplace = True)
    
```

```python
# to confirm there is no missing values in numerical variables
print(X_train[vars_with_na].isnull().sum())
print()
print(X_test[vars_with_na].isnull().sum())
```

```python
# check that we have the added binary flag column
X_train[['LotFrontage_na', 'MasVnrArea_na', 'GarageYrBlt_na']].head()
```

### 4. Temporal Variables

We need calculate the elapsed time between temporal variable and year house sold.

```python
def elapsed_years(df, var):
    # calcualte difference betwen year varialbe and year house sold
    df[var] = df['YrSold'] - df[var]
    return df
```

```python
for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    X_train = elapsed_years(X_train, var)
    X_test = elapsed_years(X_test, var)
```

```python
# check if there is a missing value in temporal variable
print([vr for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt'] if X_train[var].isnull().sum()>0])
print([vr for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt'] if X_test[var].isnull().sum()>0])
```

### 5. Numerical Variable
We have to log transform the numerical variables that do not contain zeors that we have idenfied during data analysis phase.

```python
for var in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:
    X_train[var] = np.log(X_train[var])
    X_test[var]= np.log(X_test[var])
```

```python
# check that test set does not contain null values in the engineered variables
print([var for var in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'] if X_train[var].isnull().sum()>0])
print([var for var in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'] if X_test[var].isnull().sum()>0])
```

### 6. Categorical Variables

First, we will remove the those categories within the varibles that are present in less than 1% of the observation.

```python
# let's capture the categorical variables first
cat_vars = [var for var in X_train.columns if X_train[var].dtype == 'O']
```

```python
def find_frequent_labels(df, var, rare_perc):
    # fint the labels that present more than given percentage
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp > rare_perc].index
```

```python
for var in cat_vars:
    frequent_ls = find_frequent_labels(X_train, var, 0.01)
    X_train[var] = np.where(X_train[var].isin(frequent_ls), X_train[var], 'Rare')
    X_test[var] = np.where(X_test[var].isin(frequent_ls),X_test[var], 'Rare')
```

Next, we need to transform the strings of these variables into numbers, inorder to feed data into model. 

```python

def replace_categories(train, test, var, target):
    ordered_labels = train.groupby([var])[target].mean().sort_values().index
    ordinal_label = {k:i for i, k in enumerate(ordered_labels, 0)} 
    train[var] = train[var].map(ordinal_label)
    test[var] = test[var].map(ordinal_label)
```

```python
for var in cat_vars:
    replace_categories(X_train, X_test, var, 'SalePrice')
```

### 6.Feature Scaling
Inorder to use linear models, feautures need to be scaled or normalized.

```python
train_vars = [var for var in X_train.columns if var not in ['Id', 'SalePrice']]
len(train_vars)
```

```python
X_train[['Id', 'SalePrice']].reset_index(drop=True)
```

```python
# fit scaler
scaler = MinMaxScaler() # create an instance
scaler.fit(X_train[train_vars]) #  fit  the scaler to the train set for later use

# transform the train and test set, and add on the Id and SalePrice variables
X_train = pd.concat([X_train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_train[train_vars]), columns=train_vars)],
                    axis=1)

X_test = pd.concat([X_test[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_test[train_vars]), columns=train_vars)],
                    axis=1)
```
